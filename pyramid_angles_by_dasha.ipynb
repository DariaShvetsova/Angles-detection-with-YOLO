```python
from scipy import optimize
import numpy as np
import math

from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw
from PIL.Image import core as _imaging
# import PyQt5

from PIL import ImageGrab
```

```python
data=np.array([[374.7683,  58.5940],
         [344.7909, 214.0545],
         [400.0717, 226.2337],
         [349.3854, 173.9303],
         [397.1552, 187.0602]])

top=data[0]
bp1=data[1]
bp2=data[2]
bp3=data[3]
bp4=data[4]

v1=bp1-top
v4=bp4-top

v2=bp2-top
v3=bp3-top


normalized_v1 =v1 / np.linalg.norm(v1)
normalized_v4 =v4 / np.linalg.norm(v4)

normalized_v2 =v2 / np.linalg.norm(v2)
normalized_v3 =v3 / np.linalg.norm(v3)


dot_v1_v4 = np.dot(normalized_v1, normalized_v4)

dot_v2_v3 = np.dot(normalized_v2, normalized_v3)

angle_rad14=np.arccos(dot_v1_v4)

angle_rad23=np.arccos(dot_v2_v3)

angle_grad14=np.degrees(angle_rad14)
angle_grad23=np.degrees(angle_rad23)

alpha=round(((angle_grad14+angle_grad23)/4), 2)

##############

image_path = "39.jpg"

print(image_path)


img = Image.open("39.jpg")
draw = ImageDraw.Draw(img)

draw.text((550,60),"ANGLE DETECTED!!! real alpha = {} grad".format(alpha),(255,36,0))

display(img)





```
